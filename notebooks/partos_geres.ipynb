```python
import pandas as pd

DATADIR = '../app/datasets/inputs'
DATAOUT = '../app/datasets/outputs'

geres_cols = {
    '2601 Recife': 'I GERES - Recife',
    '2602 Limoeiro': 'II GERES - Limoeiro',
    '2603 Palmares': 'III GERES - Palmares',
    '2604 Caruaru': 'IV GERES - Caruaru',
    '2605 Garanhuns': 'V GERES - Garanhuns',
    '2606 Arcoverde': 'VI GERES - Arcoverde',
    '2607 Salgueiro': 'VII GERES - Salgueiro',
    '2608 Petrolina': 'VIII GERES - Petrolina',
    '2609 Ouricuri': 'IX GERES - Ouricuri',
    '2610 Afogados da Ingazeira': 'X GERES - Afogados da Ingazeira',
    '2611 Serra Talhada': 'XI GERES - Serra Talhada',
    '2612 Goiana': 'XII GERES - Goiana'
}
```

```python
df_partos_geres_normal = pd.read_csv(
    f'{DATADIR}/partos_normal_cesario_geres_ocorr_geres_resid_2020_2022.csv', 
    sep=';'
    ).assign(TIPO_PARTO=('Normal/Cesário'))

df_partos_geres_risco = pd.read_csv(
    f'{DATADIR}/partos_normal_cesario_risco_geres_ocorr_geres_resid_2020_2022.csv', 
    sep=';'
    ).assign(TIPO_PARTO=('De Risco'))  

df_partos_geres = pd.concat([df_partos_geres_normal, df_partos_geres_risco]).\
    rename(columns=geres_cols)

df_partos_geres = df_partos_geres.\
    assign(geres_internacao=(df_partos_geres['GERES-Internação'].map(geres_cols))).\
        drop('GERES-Internação', axis=1)

df_partos_geres.head()
```

```python
df_partos_geres.to_parquet(f'{DATAOUT}/partos_por_geres_2019_2022.parquet.gzip')
```
